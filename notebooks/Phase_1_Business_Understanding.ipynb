### Task 1: Determine Business Objectives

The first objective of the data analyst is to thoroughly understand, from a business perspective, what the customer really wants to accomplish. This involves:

- **Background:** Identify the organisational context and the business area the project relates to.
- **Business Objectives:** Clearly state the primary question or problem the business wants to solve.
- **Business Success Criteria:** Define the measurable criteria that determine whether the project's outcome is considered a success from the business's point of view (e.g., "reduce customer churn by 15%", "improve prediction accuracy to 90%").

**Instructions:** In the cells below, define your business objectives, background context, and success criteria as Python variables (strings or dictionaries). This ensures your business logic is documented directly in the notebook's code state.

```python
# Defining the core project goals.
business_objectives = {
    "background": "This project is part of a larger effort in Agriculture Analytics. I'm focusing on crop yield prediction—specifically using data to figure out how to make farming more efficient and help with food supply planning.",
    "primary_objective": (
        "First, I want to see if we can forecast Rice and Wheat yields about 3 months before harvest. If we can do that using weather and fertilizer data, it'll make planning storage and transport much easier. "
        "Second, I'm looking into 'drought resilience' to see which crops handle low rainfall years with the most stable yields."
    ),
    "success_criteria": [
        "Build a regression model that can accurately predict production per hectare.",
        "Keep the prediction errors low enough that the results are actually useful for planning.",
        "Answer the research question with a clear comparison of how different crops handle dry years."
    ]
}
```

```python
# Display the defined objectives in a readable format
for key, value in business_objectives.items():
    print(f">> {key.upper().replace('_', ' ')}:")
    if isinstance(value, list):
        for item in value:
            print(f"   - {item}")
    else:
        print(f"   {value}")
    print()
```

### Task 3: Determine Data Mining Goals

A business goal states objectives in business terminology. A data mining goal states project objectives in technical terms. This task converts the business objectives identified in Task 1 into specific, measurable data mining goals:

- **Data Mining Problem Type:** Classify the problem (e.g., classification, regression, clustering, anomaly detection, association rule mining, etc.).
- **Data Mining Goals:** State the technical objectives clearly (e.g., "Build a classification model to predict [target variable] with an F1-score ≥ 0.85").
- **Data Mining Success Criteria:** Define the technical metrics that will be used to evaluate the model's performance (e.g., accuracy, precision, recall, RMSE, silhouette score).

**Instructions:** Map your business objectives to concrete data mining goals.

```python
# What I want to achieve technically.
data_mining_goals = {
    "problem_type": "Regression",
    "target": "yield_per_hectare",
    "goals": [
        "Build a model that can predict yield_per_hectare for Rice and Wheat.",
        "Find the actual correlations between rainfall, fertilizer, and how much the yields vary."
    ],
    "metrics": ["MAE (Mean Absolute Error)", "RMSE (Root Mean Squared Error)", "R-squared (R2 Score)"]
}
```

```python
# Display the mapping from business objectives to data mining goals
print("=" * 60)
print("BUSINESS TO DATA MINING GOAL MAPPING")
print("=" * 60)
print(f"\nBusiness Objective : {business_objectives['primary_objective']}")
print(f"Problem Type       : {data_mining_goals['problem_type']}")
print(f"Target Variable    : {data_mining_goals['target']}")
print(f"\nTechnical Goals:")
for goal in data_mining_goals['goals']:
    print(f"  - {goal}")
print(f"\nSuccess Metrics:")
for metric in data_mining_goals['metrics']:
    print(f"  - {metric}")
```

```python
# The roadmap for the project.
project_plan = {
    "tools": ["Pandas", "NumPy", "Matplotlib", "Seaborn", "Scikit-learn", "BeautifulSoup"],
    "techniques": ["Linear Regression", "Random Forest", "Web Scraping", "Pipeline-based Modeling"],
    "steps": [
        {"phase": "Data Understanding", "status": "In Progress"},
        {"phase": "Data Preparation", "status": "Upcoming"},
        {"phase": "Modeling", "status": "Upcoming"},
        {"phase": "Evaluation", "status": "Upcoming"},
        {"phase": "Deployment", "status": "Upcoming"}
    ],
    "justification": "Pandas is perfect for this size of data, but I'll also discuss how I'd switch to Spark if we had to handle 10TB+ of farm records."
}
```
